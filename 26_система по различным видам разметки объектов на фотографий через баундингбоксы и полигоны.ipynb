# Задача 1 Напишите функцию, которая на вход принимает путь к текстовому файлу и текст, который нужно добавить в этот файл в конце имеющихся там данных.

```python
def append_text_to_file(file_path, text):
    """
    Функция добавляет указанный текст в конец текстового файла.

    :param file_path: Путь к текстовому файлу.
    :param text: Текст, который нужно добавить.
    """
    try:
        with open(file_path, 'a', encoding='utf-8') as file:
            file.write(text + '\n')
        print(f"Текст успешно добавлен в файл {file_path}")
    except Exception as e:
        print(f"Произошла ошибка при добавлении текста в файл: {e}")

# Пример использования
file_path = 'example.txt'
text = 'Это текст, который нужно добавить в файл.'
append_text_to_file(file_path, text)
```

# Задача 2 Переместите все файлы формата .csv из директории /content/sample_data в новую /content/sample_data/all_csv

```python
import os
import shutil

# Исходная директория
source_dir = '/content/sample_data'
# Целевая директория
target_dir = '/content/sample_data/all_csv'

# Создаем целевую директорию, если она не существует
if not os.path.exists(target_dir):
    os.makedirs(target_dir)
    print(f"Создана директория: {target_dir}")

# Проходим по всем файлам в исходной директории
for filename in os.listdir(source_dir):
    # Проверяем, является ли файл формата .csv
    if filename.endswith('.csv'):
        source_file = os.path.join(source_dir, filename)
        target_file = os.path.join(target_dir, filename)
        # Перемещаем файл
        shutil.move(source_file, target_file)
        print(f"Файл {filename} перемещен в {target_dir}")

print("Все файлы формата .csv успешно перемещены.")
```

# Задача 4 Напишите функцию, которая на вход принимает путь к какой-либо директории, а на выходе мы должны получить количество папок в выбранной директории и количество файлов (отдельно).

```python
import os

def count_dirs_and_files(directory_path):
    # Инициализируем счётчики для папок и файлов
    num_dirs = 0
    num_files = 0

    # Проходим по всем элементам в указанной директории
    for item in os.listdir(directory_path):
        item_path = os.path.join(directory_path, item)

        # Проверяем, является ли элемент папкой или файлом и увеличиваем соответствующий счётчик
        if os.path.isdir(item_path):
            num_dirs += 1
        elif os.path.isfile(item_path):
            num_files += 1

    return num_dirs, num_files

# Пример использования функции
directory_path = '/content/sample_data'
num_dirs, num_files = count_dirs_and_files(directory_path)
print(f"Количество папок: {num_dirs}")
print(f"Количество файлов: {num_files}")
```

# Задача 6 На занятии мы написали функцию check_markup(), в которой мы отрисовываем ограничивающие рамки на изображении цветом (100, 0, 255). Измените код функции так, чтобы цвет рамки зависел от класса объекта (класс 0 - зеленый, класс 1 - красный).

```python
import gdown
gdown.download('https://drive.google.com/uc?id=1hbbVmNgeAXvHPdhwTSLzE7Fb4OqtABRw', None, quiet=False)
gdown.download('https://drive.google.com/uc?id=191RRwBCI0ylnYTKxfaioXiPHN3WEwq5n', None, quiet=False)
```

```python
import cv2
from google.colab.patches import cv2_imshow

def check_markup(path_to_img, path_to_txt):
    # загрузка изображения
    img = cv2.imread(path_to_img)
    img = cv2.resize(img, (640, 393))

    # загрузка разметки
    with open(path_to_txt) as f:
        annotation = f.read()

    # проверка есть ли перенос строки в последней строке
    if annotation.endswith('\n'):
        annotation = annotation.split('\n')[:-1]
    else:
        annotation = annotation.split('\n')

    # ширина и высота изображения
    width, height = img.shape[1], img.shape[0]

    # проходим по всем объектам разметки
    for obj in annotation:
        # получаем класс объекта
        obj_class = int(obj.split(' ')[0])

        # переведем относительные значения к пикселям
        Xc = int(float(obj.split(' ')[1]) * width)
        Yc = int(float(obj.split(' ')[2]) * height)
        W = int(float(obj.split(' ')[3]) * width)
        H = int(float(obj.split(' ')[4]) * height)

        # переведем значения XcYcWH в Xmin, Xmax, Ymin, Ymax
        Xmin, Xmax, Ymin, Ymax = int(Xc - W / 2), int(Xc + W / 2), int(Yc - H / 2), int(Yc + H / 2)

        # устанавливаем цвет рамки в зависимости от класса объекта
        if obj_class == 0:
            color = (0, 255, 0)  # зеленый
        elif obj_class == 1:
            color = (0, 0, 255)  # красный
        else:
            color = (100, 0, 255)  # цвет по умолчанию

        # добавляем прямоугольник на изображение
        imageRect = cv2.rectangle(img,
                                  pt1=(Xmin, Ymin),
                                  pt2=(Xmax, Ymax),
                                  color=color,
                                  thickness=2)
    # выводим результат на экран
    cv2_imshow(imageRect)

# проверим работу функции
check_markup('/content/pets.jpg', '/content/pets.txt')
```

# Задача 7 Напишите функцию, которая на вход будет принимать изображение и файл с разметкой формата Pascal VOC, а на выходе будем получать отрисованное изображение с ограничивающими рамками.

```python
gdown.download('https://drive.google.com/uc?id=1ZXUYwKD420ATmlbdGh6KrUNaTMjU6Ha-', None, quiet=False)
```

```python
import cv2
import xml.etree.ElementTree as ET
from google.colab.patches import cv2_imshow

def parse_voc_annotation(path_to_txt):
    tree = ET.parse(path_to_txt)
    root = tree.getroot()
    objects = []

    size = root.find('size')
    width = int(size.find('width').text)
    height = int(size.find('height').text)

    for obj in root.findall('object'):
        obj_name = obj.find('name').text
        bbox = obj.find('bndbox')
        xmin = int(bbox.find('xmin').text)
        ymin = int(bbox.find('ymin').text)
        xmax = int(bbox.find('xmax').text)
        ymax = int(bbox.find('ymax').text)
        objects.append((obj_name, xmin, ymin, xmax, ymax))

    return objects, width, height

def check_markup_voc(path_to_img, path_to_txt):
    # загрузка изображения
    img = cv2.imread(path_to_img)

    # парсинг разметки Pascal VOC
    objects, orig_width, orig_height = parse_voc_annotation(path_to_txt)

    # новый размер изображения
    new_width, new_height = 640, 393
    img = cv2.resize(img, (new_width, new_height))

    # цвета для каждого класса
    colors = {
        'cat': (0, 255, 0),  # зеленый для класса 'cat'
        'dog': (0, 0, 255)   # красный для класса 'dog'
    }

    # проходим по всем объектам разметки
    for obj in objects:
        obj_name, xmin, ymin, xmax, ymax = obj

        # масштабируем координаты
        xmin = int(xmin * new_width / orig_width)
        ymin = int(ymin * new_height / orig_height)
        xmax = int(xmax * new_width / orig_width)
        ymax = int(ymax * new_height / orig_height)

        color = colors.get(obj_name, (100, 0, 255))  # цвет по умолчанию, если класс не 'cat' и не 'dog'

        # добавляем прямоугольник на изображение
        imageRect = cv2.rectangle(img,
                                  pt1=(xmin, ymin),
                                  pt2=(xmax, ymax),
                                  color=color,
                                  thickness=2)
        # добавляем текст на изображение
        cv2.putText(img, obj_name, (xmin, ymin - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, color, 2)

    # выводим результат на экран
    cv2_imshow(imageRect)

# поверим работу функции
check_markup_voc('/content/pets.jpg', '/content/pets.xml')
```

# Задача 8 Напишите функцию, которая на вход будет принимать изображение и файл с разметкой формата COCO, а на выходе будем получать отрисованное изображение с полигонами.

```python
import gdown

gdown.download('https://drive.google.com/uc?id=10g9Fry4uv2CUONHzrxWtuaXTic3Ganb4', None, quiet=False)
```

```python
import cv2
import json
import numpy as np
from google.colab.patches import cv2_imshow

def draw_polygons_coco(path_to_img, path_to_json):
    # чтение изображения
    img = cv2.imread(path_to_img)

    # чтение json файла
    with open(path_to_json, 'r') as f:
        coco = json.load(f)

    # проходим по всем объектам
    for ann in coco['annotations']:
        # переводим значения из строковых в целочисленные и
        # получаем список xy координат точек полигонов
        coords_list = [int(float(i)) for i in ann['segmentation'][0]]

        # меняем размерность соответствующую функции cv2.polylines()
        polygons = np.reshape(coords_list, (1, len(coords_list) // 2, 2))

        # добавляем полигон на изображение
        img = cv2.polylines(img,
                            pts=polygons,
                            isClosed=True,
                            color=(230, 50, 10),
                            thickness=2)

    # выводим результат на экран
    cv2_imshow(img)

# Пример использования функции
path_to_img = '/content/pets.jpg'
path_to_json = '/content/segment_annotations.coco.json'  # замените на ваш путь к файлу разметки

draw_polygons_coco(path_to_img, path_to_json)
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from PIL import Image

# Путь к изображению и файлу разметки
path_to_img = '/content/pets.jpg'
path_to_txt = '/content/pets.txt'

# Загрузка изображения
img = Image.open(path_to_img)
width, height = img.size

# Чтение разметки YOLO
with open(path_to_txt, 'r') as f:
    annotations = f.readlines()

# Создание графика
fig, ax = plt.subplots(1)
ax.imshow(img)

# Проходим по всем объектам разметки
for ann in annotations:
    # Парсинг строки аннотации
    class_id, x_center, y_center, w, h = map(float, ann.strip().split())
    x_center, y_center, w, h = x_center * width, y_center * height, w * width, h * height

    # Вычисляем координаты верхнего левого угла
    xmin = x_center - w / 2
    ymin = y_center - h / 2

    # Добавляем прямоугольник на изображение
    rect = patches.Rectangle((xmin, ymin), w, h, linewidth=2, edgecolor='r', facecolor='none')
    ax.add_patch(rect)

# Отображение результата
plt.show()
```

# Задача 10 Напишите функцию, которая на вход принимает файл json с разметкой в формате COCO (положение ограничивающих рамок объектов), меняет разметку на формат YOLO и сохраняет результаты в файлах txt для каждого изображения.

```python
import gdown

gdown.download('https://drive.google.com/uc?id=1cr1vQawV3cAIARbNmcBu1URTc2ngon3U', None, quiet=False)
```

```python
import json
from pprint import pprint

# чтение json файла
with open('/content/annotations.coco.json', 'r') as f:
    coco = json.load(f)
pprint(coco)
```

```python
import json
import os
import glob

def coco_to_yolo(coco_json_path, output_dir):
    # Создаем директорию, если она не существует
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    # Читаем json файл с разметкой COCO
    with open(coco_json_path, 'r') as f:
        coco = json.load(f)

    # Получаем размеры изображений
    image_data = {img['id']: (img['width'], img['height'], img['file_name']) for img in coco['images']}

    # Словарь для хранения аннотаций для каждого изображения
    annotations = {img_id: [] for img_id in image_data.keys()}

    # Проходим по всем объектам разметки
    for ann in coco['annotations']:
        image_id = ann['image_id']
        class_id = ann['category_id']
        bbox = ann['bbox']

        # Получаем ширину и высоту изображения
        width, height, file_name = image_data[image_id]

        # Переводим координаты из COCO в YOLO формат
        x, y, w, h = bbox
        x_center = x + w / 2
        y_center = y + h / 2

        # Нормализуем координаты
        x_center /= width
        y_center /= height
        w /= width
        h /= height

        # Формируем строку разметки в формате YOLO
        yolo_annotation = f"{class_id} {x_center:.6f} {y_center:.6f} {w:.6f} {h:.6f}\n"

        # Добавляем аннотацию в словарь
        annotations[image_id].append(yolo_annotation)

    # Записываем результаты в файлы .txt для каждого изображения
    for image_id, yolo_annotations in annotations.items():
        _, _, file_name = image_data[image_id]
        output_file_path = os.path.join(output_dir, f"{os.path.splitext(file_name)[0]}.txt")
        with open(output_file_path, 'w') as output_file:
            output_file.writelines(yolo_annotations)

# Пример использования функции
coco_json_path = '/content/annotations.coco.json'
output_dir = '/content/yolo_annotations'
coco_to_yolo(coco_json_path, output_dir)

# Проверка наличия созданных файлов .txt в папке output_dir
created_files = glob.glob(os.path.join(output_dir, '*.txt'))
print("Созданные файлы:")
for file in created_files:
    print(file)
```
