# Задание 1

Напишите программу по оценке экзаменационных результатов пользователей на основе введенных баллов (максимум 100 баллов).

1. Если пользователь набрал 90 баллов и выше - получает оценку "Отлично".
2. Оценка "Хорошо" начинается от 70 баллов (включительно).
3. Оценка "Удовлетворительно" от 50 баллов (включительно).
4. Если оценка ниже - то экзамен считается проваленным, предложите пользователю прийти на пересдачу.

```python
bali = int(input('Чтобы узнать оценку экзаменационных результатов введите колличество полученных балов:\n', ))
if bali >= 90:
  print('Ваша оценка: Отлично!')
elif bali >= 70:
  print('Ваша оценка: Хорошо!')
elif bali >= 50:
  print('Ваша оценка: Удовлетворительно!')
else:
  print('К сожалению вы не сдали экзамен: Приходите на перездачу!')
```

# Задание 2

Машины по очереди заезжают на парковку, пока в ней есть свободные места. Вместимость парковки 8 авто.

Используя цикл `while`, каждый заезд очередной машины сопроводите сообщением `"Машина заехала"`. Выведите на экран количество свободных мест на парковке после заезда очередной машины.

Когда парковка окажется полной - сообщите остальным автовладельцам, что места в ней больше нет

```python
mst = 0
while mst <= 8:
  if mst < 8:
    print(f'количество свободных мест на парковке: {8 - mst}')
  else:
    print('К сожалению свободные места на парковке закончились')
  mst +=1
```
